# words and sentences count ‚öæ

```python
# Download the punkt tokenizer
nltk.download("punkt")

# Define a function to count words and sentences
def count_words_sentences(file_name):
    # Open the file in read mode
    with open(file_name, "r") as file:
        # Read the file content
        content = file.read()
        # Use nltk.word_tokenize to tokenize the content into words
        words = nltk.word_tokenize(content)
        # Use nltk.sent_tokenize to tokenize the content into sentences
        sentences = nltk.sent_tokenize(content)
        # Return the length of words and sentences lists
        return len(words), len(sentences)

```

```python
# Call the function with a file name
words_count, sentences_count = count_words_sentences("/content/output.txt")

# Print the results
print(f"The file has {words_count} words and {sentences_count} sentences.")
```

#Split into sentences and words ‚úà

```python
# Define a function to print word counts
def words(file_name,outfileName):
    # Open the file in read mode
    with open(file_name, "r") as file:
        # Read the file content
        content = file.read()
        # Use nltk.word_tokenize to tokenize the content into words
        words = nltk.word_tokenize(content)
        # Use a list comprehension to filter out dots and question marks
        words = [word for word in words if word not in [".", "?",",","(",")","[","]","the","of",'in','was','at','and','to','is','a','He','an',
                                                        'as','for','‚Äì','he','from','his','on','with','\'s','In',':',';','``','The','""','"','that','also','by']]
        
        with open(outfileName , "w") as text_file:
              for word in words:
                text_file.write(word + ",")
        print(f"number of words of file : {file_name} => {len(words)}")
        # Use nltk.FreqDist to create a frequency distribution of words
        return words

mathWords = words("/content/MathSummary.txt","MathWordbyWord.txt")
politicsWords = words("/content/PoliticsSummary.txt",'politicsWordbyWord.txt')
common_words = set(mathWords) & set(politicsWords)
print ( 20 * "=")
print(f"number of common words : {len(common_words)}")
#print ( 20 * "=")
#print(common_words)
with open("common_words.txt",'w') as txt_file:
  for word in common_words:
    txt_file.write(word + ",")
```

```python
def Sentences(inputfile , outputfile):
  with open(inputfile) as f:
      text = f.read()
  counter = 0
  sentences = nltk.sent_tokenize(text)
  with open(outputfile,'w') as text_file:
    for sentence in sentences:
      text_file.write(sentence + "\n")
      counter += 1
  print(f"number of sentences of file : {inputfile} => {counter}")

Sentences("/content/PoliticsSummary.txt","politicsSentbySent.txt")
Sentences("/content/MathSummary.txt","mathSentbySent.txt")
      
```

#number of distinct worlds ‚öì

```python
# Define a function to print word counts
def print_word_counts(file_name):
    distinct_words_count = {}
    # Open the file in read mode
    with open(file_name, "r") as file:
        # Read the file content
        content = file.read()
        # Use nltk.word_tokenize to tokenize the content into words
        words = nltk.word_tokenize(content)
        # Use a list comprehension to filter out dots and question marks
        words = [word for word in words if word not in [".", "?",",","(",")","[","]","the","of",'in','was','at','and','to','is','a','He','an',
                                                        'as','for','‚Äì','he','from','his','on','with','\'s','In',':',';','``','The','""','"','that','also','by']]
        # Use nltk.FreqDist to create a frequency distribution of words
        freq_dist = nltk.FreqDist(words)
        # Use sorted to sort the frequency distribution by word count in descending order
        sorted_freq_dist = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
        # Loop through each word and its count in the sorted frequency distribution
        for word, count in sorted_freq_dist:
            # Print the word and its count
            distinct_words_count[word] = count
            #print(f"{word}: {count}")
        return distinct_words_count

# Call the function with a file name
words_count = print_word_counts("/content/output.txt")
print(len(words_count))

```
